```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import blob_doh
from skimage.io import imread
from skimage.color import rgb2gray
from google.colab import files

# Function to detect blobs using DoH (Determinant of Hessian)
def detect_blobs(gray):
    # Adjusted parameters for DoH to detect more blobs
    blobs = blob_doh(gray, max_sigma=50, threshold=0.005, min_sigma=1)
    return blobs

# Function to plot DoH blobs
def plot_blobs(image, gray, title):
    blobs = detect_blobs(gray)  # Only use DoH here
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.imshow(image)
    for blob in blobs:
        y, x, r = blob
        ax.add_patch(plt.Circle((x, y), r, color='lime', linewidth=1.5, fill=False))
    ax.set_title(f"{title} - DoH: {len(blobs)} blobs")
    ax.axis('off')
    plt.show()

# Upload and process platelets image
# Upload and process platelets image
print("Upload light microscope image (platelets)")
uploaded1 = files.upload()
image_path1 = next(iter(uploaded1))
image1 = imread(image_path1)

# Convert the image to RGB if it has an alpha channel
if image1.shape[2] == 4:  # Check if image has 4 channels (RGBA)
    image1 = image1[:, :, :3]  # Slice the array to keep only the first 3 channels (RGB)

gray1 = rgb2gray(image1)

# Plot blobs on platelets image using DoH
plot_blobs(image1, gray1, "Platelets")

# Upload and process candy image
print("Upload candy.jpg")
uploaded2 = files.upload()
image_path2 = next(iter(uploaded2))
candy = cv2.imread(image_path2, cv2.IMREAD_GRAYSCALE)

# Plot blobs on candy image using DoH
plot_blobs(candy, candy, "Candy")

# Upload and process sat1 image
print("Upload sat1.jpg")
uploaded3 = files.upload()
sat1_path = next(iter(uploaded3))
sat1 = imread(sat1_path)

# Convert the image to RGB if it has an alpha channel
if sat1.shape[2] == 4:  # Check if image has 4 channels (RGBA)
    sat1 = sat1[:, :, :3]  # Slice the array to keep only the first 3 channels (RGB)

sat1_gray = rgb2gray(sat1)

# Plot blobs on sat1 image using DoH
plot_blobs(sat1, sat1_gray, "SAT1")

# Upload and process sat2 image
print("Upload sat2.jpg")
uploaded4 = files.upload()
sat2_path = next(iter(uploaded4))
sat2 = imread(sat2_path)

# Check if sat2 has an alpha channel and convert to RGB if necessary
if sat2.shape[2] == 4:  # Check if image has 4 channels (RGBA)
    sat2 = sat2[:, :, :3]  # Slice the array to keep only the first 3 channels (RGB)

sat2_gray = rgb2gray(sat2)

# Plot blobs on sat2 image using DoH
plot_blobs(sat2, sat2_gray, "SAT2")

```
